# Random Forest
- Decision Tree는 Overfitting 가능성이 높다는 약점이 있다. Random Forest는 여러 개의 의사결정나무(Decision Tree)를 조합하여 예측을 수행하는 앙상블 모델로, 단일 트리보다 일반화 성능이 우수하다.
- 앙상블(Ensemble)이란 강력한 하나의 모델을 사용하는 대신 보다 약한 여러개의 모델을 조합하여 예측하는 방법이다. (random forestm, xgboost 등)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import randint
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn import preprocessing
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, RandomizedSearchCV
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import classification_report
from IPython.display import Image, display

# %matplotlib inline
```

## 2. Preprocessing

### 2-1. null 제거 -> 없음

```python
boston_df.dropna(how='any', inplace=True)
boston_df.shape
```

- 트리계열 모델은 비선형 모델로 분포에 영향을 크게 받지 않기 때문에 log 변환 등의 값 변환이 없어도 괜찮을 수 있다?

### 2-2. X,y 분할

```python
X = boston_df.drop('medv', axis=1)
y = boston_df['medv']
print(X.shape, y.shape)
```

### 2-3. train test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

## 3. Modeling

```python
display(Image(filename='../image/random_forest_hyperparameter_info.png', width=600))
```

```python
rf_reg = RandomForestRegressor(n_estimators=8,
                               max_features=6,
                               random_state=123)

rf_reg.fit(X_train, y_train)

y_pred = rf_reg.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Test RMSE:", round(rmse, 2))
print("Test R²:", round(r2, 4))
```

### 3-1. 하이퍼파라미터 튜닝

```python
param_distributions = {
    'n_estimators': randint(2, 500),
    'max_depth': randint(5, 20),
    'min_samples_split': randint(2, 10),
    'min_samples_leaf': randint(1, 5),
    'max_features': ['auto', 'sqrt', 'log2']
}

model = RandomForestRegressor(random_state=42)

random_search = RandomizedSearchCV(
    estimator=model,
    param_distributions=param_distributions,
    n_iter=30,               # 30번 샘플링해서 모델 학습
    scoring='neg_mean_squared_error',
    cv=5,
    verbose=1,
    random_state=42,
    n_jobs=-1
)

random_search.fit(X_train, y_train)

best_model = random_search.best_estimator_
print("Best params:", random_search.best_params_)
```

```python
best_model = random_search.best_estimator_
```

```python
y_pred = best_model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Test RMSE:", round(rmse, 2))
print("Test R²:", round(r2, 4))
```
